```python
import matplotlib.pyplot as plt
```

```python
# Simple plot, x and y values are given as lists
plt.plot([1,2,3,4,5],[1,4,9,16,25])
# add title to the plot
plt.title('Simple Plot')
# add x and y labels to the plot
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
# add grid to the plot
plt.grid(True)
# show the plot
plt.show()
```
